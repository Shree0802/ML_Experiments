# **Introduction of NumPy.**


***NumPy :- Numerical Python***

**Introduction :-**

NumPy stands for Numerical Python. It is an open-source Python library used for numerical and scientific computing. NumPy provides a powerful multidimensional array object called ndarray, along with a large collection of mathematical functions to efficiently perform operations on arrays.

NumPy is one of the most important libraries in Python because many other libraries such as Pandas, SciPy, Matplotlib, Scikit-learn, and TensorFlow are built on top of it.

# **History of NumPy :-**

NumPy has a rich history that dates back to the early development of scientific computing in Python.

<br>

**Early Development**

Before NumPy existed, Python had limited support for numerical and scientific computations. To address this issue, two separate libraries were developed:

Numeric (released in 1995): The first major Python library for numerical computing.
Numarray (released in 2001): Designed to handle very large arrays more efficiently than Numeric.

<br>

**Creation of NumPy**

In 2005, Travis Oliphant, an American software developer and scientist, created NumPy by combining the best features of Numeric and Numarray into a single, unified library.

His goal was to provide:

- A faster numerical computing library
- Better memory efficiency
- A single standard array package for Python
- Compatibility with scientific applications

# **Why NumPy was Developed?**

## **Need for NumPy**

Python is a simple and powerful programming language, but its built-in data structures, such as lists, are not designed for high-performance numerical computing. As the amount of data increases, operations on Python lists become slower and consume more memory.

NumPy was developed to overcome these limitations and provide a faster, more efficient way to perform numerical and scientific computations.

## **Problems with Python Lists**

Python lists have several limitations when working with numerical data:

- Slower execution for mathematical operations.
- Higher memory consumption.

## **Why NumPy is Better**

- NumPy solves these problems by providing a powerful multidimensional array object called **ndarray**.
- These arrays are stored in continuous memory locations, making computations much faster and more memory efficient than Python lists.



## **Applications of NumPy :-**

NumPy is widely used in various fields that require fast numerical computations and efficient data processing. Some major applications of NumPy are:

- Data Science: Used for data analysis, preprocessing, and manipulation.
- Machine Learning: Serves as the foundation for libraries such as Scikit-learn, TensorFlow, and PyTorch.
- Artificial Intelligence: Performs numerical computations required for AI algorithms.
- Deep Learning: Handles tensors, matrices, and mathematical operations used in neural networks.
- Scientific Computing: Solves complex mathematical and scientific problems.
- Image Processing: Represents images as arrays and performs operations like filtering and transformation.
- Computer Vision: Used in object detection, image recognition, and video processing.
- Signal Processing: Processes audio signals, digital signals, and communication data.

## **Features of NumPy :-**


- Multidimensional Arrays: NumPy supports powerful N-dimensional arrays using the ndarray object.
- High Performance: NumPy performs numerical operations faster than standard Python lists.
- Memory Efficient: NumPy arrays use less memory for storing large amounts of numerical data.
- Vectorized Operations: Mathematical operations can be performed on entire arrays without writing loops.
- Broadcasting: NumPy can perform operations on arrays of different compatible shapes.
- Mathematical Functions: It provides many built-in mathematical and statistical functions.
- Indexing and Slicing: NumPy provides efficient methods to access and modify array elements.
- Linear Algebra: It supports matrix operations, determinants, inverses, and other linear algebra operations.

### **Creating NumPy Arrays :-**

A NumPy array is a collection of elements stored in a structured format. The main array object in NumPy is called **ndarray (N-dimensional array)**.

NumPy arrays can be one-dimensional, two-dimensional, three-dimensional, or multidimensional.



The `np.array()` function is used to create a NumPy array from a Python list or tuple.

#### **Syntax :-**

```python
np.array(object)
```

### **Types of NumPy Arrays Based on Dimensions :-**
<br>
#### **1. Zero-Dimensional Array (0D)**

A zero-dimensional array contains a single value. It is also called a scalar array.

#### **2. One-Dimensional Array (1D)**

A one-dimensional array contains elements in a single row or sequence.

Example:

`[10, 20, 30, 40]`

#### **3. Two-Dimensional Array (2D)**
**bold text**
A two-dimensional array contains rows and columns. It is similar to a matrix.

Example:

`[[1, 2, 3], [4, 5, 6]]`

#### **4. Three-Dimensional Array (3D)**

A three-dimensional array is an array containing multiple two-dimensional arrays.

#### **5. Multidimensional Array**

NumPy supports arrays with more than three dimensions. These are called multidimensional or N-dimensional arrays.

**Example :-**

````python
import numpy as np

# 0D Array
arr0 = np.array(10)
print("0D Array:")
print(arr0)

# 1D Array
arr1 = np.array([10, 20, 30, 40])
print("\n1D Array:")
print(arr1)

# 2D Array
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])
print("\n2D Array:")
print(arr2)

# 3D Array
arr3 = np.array([[[1, 2], [3, 4]],
                 [[5, 6], [7, 8]]])

print("\n3D Array:")
print(arr3)
````

## **Mathematical and Statistical Functions in NumPy :-**

NumPy provides many built-in mathematical and statistical functions to perform calculations on arrays efficiently.

These functions are useful for analyzing numerical data and performing mathematical operations.

### **Mathematical Functions :-**
Mathematical functions are used to perform calculations on NumPy arrays.

### sum()

The sum() function calculates the total sum of all elements in an array.

### min()

The min() function returns the smallest value from an array.

### max()

The max() function returns the largest value from an array.

### sqrt()

The sqrt() function calculates the square root of each element in an array.
<br>
<br>
### **Statistical Functions :-**

Statistical functions are used to analyze and summarize numerical data.

### mean()

The mean() function calculates the average of all elements in an array.

### median()

The median() function returns the middle value of the data after arranging the values in sorted order.

### std()

The std() function calculates the standard deviation of the array. Standard deviation shows how much the values are spread from the mean.

### var()

The var() function calculates the variance of the array. Variance measures the spread of data around the mean.

<br>

These functions make numerical and statistical calculations easier and faster when working with NumPy arrays.

### **Example :-**

````python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Array:", arr)

print("Sum:", np.sum(arr))
print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Variance:", np.var(arr))
print("Square Root:", np.sqrt(arr))
````

### **Joining Arrays in NumPy**

**concatenate()**

Description: concatenate() is used to join two or more arrays into a single array.



````python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

result = np.concatenate((arr1, arr2))
print(result)
````

**vstack()**

Description: vstack() is used to stack arrays vertically (row-wise).

````python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

result = np.vstack((arr1, arr2))
print(result)
````

hstack()

Description: hstack() is used to stack arrays horizontally (side by side).

````python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

result = np.hstack((arr1, arr2))
print(result)
````

### **Splitting Arrays in NumPy**

split()

Description: split() is used to divide an array into multiple equal sub-a

````python
arr = np.array([1, 2, 3, 4, 5, 6])

result = np.split(arr, 3)

print(result)
````

### **Boolean Indexing and Filtering in NumPy**

Boolean Indexing

Description: Boolean indexing is used to select elements that satisfy a given condition.

````python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print(arr[arr > 25])
````

Filtering Even Numbers

Description: Filter all even numbers from a NumPy array.

````python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

print(arr[arr % 2 == 0])
````

Multiple Conditions

Description: Apply more than one condition to filter array elements.

````python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print(arr[(arr >= 20) & (arr <= 40)])
````

### **Sorting and Searching in NumPy**
sort()

Description: sort() is used to arrange array elements in ascending order.

````python
arr = np.array([40, 10, 30, 20])

result = np.sort(arr)

print(result)
````

where()

Description: where() is used to find the indexes of elements that satisfy a given condition.

````python
arr = np.array([10, 20, 30, 40, 50])

result = np.where(arr > 25)

print(result)
````

### **Copy vs View in NumPy :-**
copy()

Description: copy() creates a new independent array, so changes in the original array do not affect the copied array.

````python
arr = np.array([1, 2, 3])

copy_arr = arr.copy()
arr[0] = 10

print(arr)
print(copy_arr)
````

view()

Description: view() creates a new view of the original array, so changes in the original array can affect the view.

````python
arr = np.array([1, 2, 3])

view_arr = arr.view()
arr[0] = 10

print(arr)
print(view_arr)
````

## **Broadcasting in NumPy :-**

Description: Broadcasting allows NumPy to perform arithmetic operations on arrays of different but compatible shapes without manually resizing them.

````python
import numpy as np

arr = np.array([10, 20, 30])

result = arr + 5

print(result)
````

#### **Broadcasting Between Two Arrays**

Description: Broadcasting automatically expands the smaller array to match the larger array during an operation.

````python
import numpy as np

arr1 = np.array([[1], [2], [3]])
arr2 = np.array([10, 20, 30])

result = arr1 + arr2

print(result)
````

## **Matrix Operations in NumPy :-**

Matrix Multiplication (*)

Description: Matrix multiplication combines two matrices to produce a new matrix (use matmul() or dot() for true matrix multiplication).

````python
arr1 = np.array([[1, 2],
                 [3, 4]])

arr2 = np.array([[5, 6],
                 [7, 8]])

print(arr1 * arr2)
````

dot()

Description: dot() performs matrix multiplication between two arrays.

````python
arr1 = np.array([[1, 2],
                 [3, 4]])

arr2 = np.array([[5, 6],
                 [7, 8]])

print(np.dot(arr1, arr2))
````

matmul()

Description: matmul() is used to perform matrix multiplication between two arrays.

````python
arr1 = np.array([[1, 2],
                 [3, 4]])

arr2 = np.array([[5, 6],
                 [7, 8]])

print(np.matmul(arr1, arr2))
````

inverse()

Description: np.linalg.inv() is used to find the inverse of a square matrix.

The inverse exists only for square matrices whose determinant is not equal to 0.

### **Handling Missing Values in NumPy :-**

np.nan

Description: np.nan represents a missing or undefined numerical value.

Example :-

````python
arr = np.array([10, 20, np.nan, 40])

print(arr)
````

isnan()

Description: isnan() checks whether each element is NaN or not.

````python
arr = np.array([10, np.nan, 30])

print(np.isnan(arr))
````

nanmean()

Description: nanmean() calculates the mean while ignoring NaN values.

````python
arr = np.array([10, 20, np.nan, 40])

print(np.nanmean(arr))
````

nansum()

Description: nansum() calculates the sum while ignoring NaN values.

````python
arr = np.array([10, 20, np.nan, 40])

print(np.nansum(arr))
````
